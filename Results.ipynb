```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import numpy as np

results = pickle.load(open('full_results','rb'))

data_sizes = range(100,11487,16)
data_sizes.append(11487)
cnn_data = {'accuracy' : results['CNN']['IID'][0]+results['CNN']['SGD_Uncertainty'][0]+results['CNN']['CNN_Uncertainty'][0],
            'training set size' : data_sizes+data_sizes+data_sizes,
            'strategy' : ['IID']*len(data_sizes)+['SGD Uncertainty']*len(data_sizes)+['CNN Uncertainty']*len(data_sizes)
            }
cnn_df = pd.DataFrame(data=cnn_data)
g = sns.factorplot(x='training set size',
                   y = 'accuracy',
                   hue = 'strategy',
                   data = cnn_df
                   )
plt.show()
```

```python
sgd_data = {'accuracy' : results['SGD']['IID'][0]+results['SGD']['SGD_Uncertainty'][0]+results['SGD']['CNN_Uncertainty'][0],
            'training set size' : data_sizes+data_sizes+data_sizes,
            'strategy' : ['IID']*len(data_sizes)+['SGD Uncertainty']*len(data_sizes)+['CNN Uncertainty']*len(data_sizes)
            }
sgd_df = pd.DataFrame(data=sgd_data)
g = sns.factorplot(x='training set size',
                   y = 'accuracy',
                   hue = 'strategy',
                   data = sgd_df
                   )
plt.show()
```

```python
results = pickle.load(open('full_results2','rb'))

avg_results = {}
for k1 in results.keys():
    avg_results[k1] = {}
    for k2 in results[k1].keys():
        avg_results[k1][k2] = np.mean(results[k1][k2], axis=0).tolist()

data_sizes = range(100,3000,20)
data_sizes.append(300)
cnn_data = {'accuracy' : avg_results['CNN']['IID']+avg_results['CNN']['SGD_Uncertainty']+avg_results['CNN']['CNN_Uncertainty'],
            'training set size' : data_sizes+data_sizes+data_sizes,
            'strategy' : ['IID']*len(data_sizes)+['SGD Uncertainty']*len(data_sizes)+['CNN Uncertainty']*len(data_sizes)
            }
cnn_df = pd.DataFrame(data=cnn_data)
g = sns.factorplot(x='training set size',
                   y = 'accuracy',
                   hue = 'strategy',
                   data = cnn_df
                   )
plt.show()
```

```python
sgd_data = {'accuracy' : avg_results['SGD']['IID']+avg_results['SGD']['SGD_Uncertainty']+avg_results['SGD']['CNN_Uncertainty'],
            'training set size' : data_sizes+data_sizes+data_sizes,
            'strategy' : ['IID']*len(data_sizes)+['SGD Uncertainty']*len(data_sizes)+['CNN Uncertainty']*len(data_sizes)
            }
sgd_df = pd.DataFrame(data=sgd_data)
g = sns.factorplot(x='training set size',
                   y = 'accuracy',
                   hue = 'strategy',
                   data = sgd_df
                   )
plt.show()
```
